```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb
# Parameters
n = 30
a, b, c = 1, 1, 5
d = n - a - b - c
def normalize(a, b):
    s = 0
    for k in range(b+1):
        s+= (-1)**(b-k) * comb(b, k) / (b-k+a+1)
    return s

# Define the function
def f(x, y):
    return  (x**a) * ((1 - x)**b) * (y**c) * ((1 - y)**(n - a - b - c))

coeff = normalize(a, b)
print(coeff)

# Generate grid
x = np.linspace(0.0, 1, 300)
y = np.linspace(0.0, 1, 300)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
m = max([max(a) for a in Z])
print(m)
Z1 = 1/m* f(X, Y)

# Plot heatmap
plt.figure(figsize=(6, 5))
plt.imshow(Z, extent=[0.2, 0.8, 0.2, 0.8], origin='lower', cmap='viridis', aspect='auto')
plt.colorbar(label='f(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Heatmap of f(x, y) = x^{a} * (1-x)^{b} * y^{c} * (1-y)^{n-a-b-c}')
plt.tight_layout()

plt.figure(figsize=(6, 5))
plt.imshow(Z1, extent=[0, 1, 0, 1], origin='lower', cmap='viridis', aspect='auto')
plt.colorbar(label='f(x, y)')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Heatmap of f(x, y) = x^{a} * (1-x)^{b} * y^{c} * (1-y)^{n-a-b-c}')
plt.tight_layout()
plt.show()
```
